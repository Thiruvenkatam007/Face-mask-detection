```python
# Importing required packages
import cv2
import pickle
```

## Used viola and jones algorithms to detect face 


```python
haar_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

Labling back the numbers to text in target column

```python
names = {1 : 'Mask', 0 : 'No mask'}
```

Loading pickled model

```python
pickled_svm = pickle.load(open('model_svm.pkl', 'rb'))
```

Detecting the mask in video (frame by frame analysis)

```python
capture = cv2.VideoCapture(1)
data = []
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag,img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h), (128,255,255),4)
            face = img[y:y+h,x:x+w,:]
            face = cv2.resize(face, (224,224))
            face = face.reshape(1,-1)
            pred = pickled_svm.predict(face)[0]
            n = names[int(pred)]
            cv2.putText(img,n,(x,y),font,1,(170,255,128),2)
            print(n)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == ord("q"):
            break
capture.release()
cv2.destroyAllWindows()
```
